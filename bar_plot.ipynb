```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_context("paper")
sns.set_style("white")
```

```python
N = 5
direct = [0.19, 0.21, 0.22, 0.22, 0.23]
indirect = [0.05, 0.04, 0.03, 0.03, 0.01]
#labels = ['6-9', '6-11', '6-10', '5-7', '7-1']
labels = ['5-8', '5-10', '5-9', '4-6', '6-0']
total = 0.25
ind = range(N)
width = 0.35

p1 = plt.bar(ind, direct, width)
p2 = plt.bar(ind, indirect, width, bottom=direct)

plt.ylabel('Effect', fontsize=16)
plt.title('Effects of top heads', fontsize=18)
plt.xticks(ind, labels, size=14)
plt.yticks(size=14)
#plt.yticks(np.arange(0, 81, 10))
p3 = plt.axhline(total, linestyle='--')
plt.legend((p3, p2[0], p1[0]), ('Total', 'Indirect', 'Direct'), loc='best', fontsize=16)
plt.figure(figsize=(10, 15))
plt.savefig('stacked.png')
plt.show()
```
